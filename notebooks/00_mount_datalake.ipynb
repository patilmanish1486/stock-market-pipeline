```python
STORAGE_ACCOUNT = "stockpipelinedatalake"
STORAGE_KEY = "YOUR_AZURE_STORAGE_KEY_HERE"

print("Credentials set!")
```

```python
dbutils.fs.mount(
    source="wasbs://bronze@stockpipelinedatalake.blob.core.windows.net",
    mount_point="/mnt/bronze",
    extra_configs={
        f"fs.azure.account.key.{STORAGE_ACCOUNT}.blob.core.windows.net": STORAGE_KEY
    }
)
print("Bronze container mounted!")
```

```python
mounts = [m.mountPoint for m in dbutils.fs.mounts()]
print("Active mounts:")
for m in mounts:
    print(f"  {m}")
```
